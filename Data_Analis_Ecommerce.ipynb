### Berapa banyak uang yang dihabiskan oleh pelanggan dalam beberapa bulan terakhir?

```python
# Meresample DataFrame berdasarkan bulan ('M') menggunakan kolom 'order_approved_at'
monthly_spend_df = all_data.resample(rule='M', on='order_approved_at').agg({
    "payment_value": "sum"
})
# Mengubah format indeks menjadi 'Bulan-Tahun'
monthly_spend_df.index = monthly_spend_df.index.strftime('%B')
# Me-reset indeks agar 'Bulan-Tahun' menjadi kolom biasa
monthly_spend_df = monthly_spend_df.reset_index()
# Mengganti nama kolom "payment_value" menjadi "total_spend"
monthly_spend_df.rename(columns={"payment_value": "total_spend"}, inplace=True)
# Menampilkan beberapa baris pertama dari DataFrame
monthly_spend_df.head()
```

```python
# Mengurutkan DataFrame berdasarkan kolom 'total_spend'
monthly_spend_df = monthly_spend_df.sort_values('total_spend')
# Menghapus duplikat berdasarkan kolom 'order_approved_at', tetapi mempertahankan yang terakhir
monthly_spend_df = monthly_spend_df.drop_duplicates('order_approved_at', keep='last')
# Mengurutkan ulang DataFrame berdasarkan 'total_spend'
monthly_spend_df = monthly_spend_df.sort_values(by='total_spend')
# Menampilkan DataFrame yang telah diurutkan ulang
monthly_spend_df
```

```python
# Membuat mapping antara nama bulan dalam bahasa Indonesia dengan format numerik
month_mapping = {
    "Januari": 1,
    "Februari": 2,
    "Maret": 3,
    "April": 4,
    "Mei": 5,
    "Juni": 6,
    "Juli": 7,
    "Augustus": 8,
    "September": 9,
    "Oktober": 10,
    "November": 11,
    "Desember": 12
}

# Mapping bulan dalam kolom "order_approved_at" ke format numerik dan menyimpannya dalam kolom baru "month_numeric"
monthly_spend_df["month_numeric"] = monthly_spend_df["order_approved_at"].map(month_mapping)

# Mengurutkan DataFrame berdasarkan "month_numeric"
monthly_spend_df = monthly_spend_df.sort_values("month_numeric")

# Menghapus kolom "month_numeric"
monthly_spend_df = monthly_spend_df.drop("month_numeric", axis=1)

# Plotting total pengeluaran konsumen per bulan dengan warna biru
plt.figure(figsize=(10, 5))
plt.plot(
    monthly_spend_df["order_approved_at"],
    monthly_spend_df["total_spend"],
    marker='o',
    linewidth=2,
    color="#3498db"  # Warna biru
)
plt.title("Total Pengeluaran Konsumen per Bulan (2018)", loc="center", fontsize=20)
plt.xticks(fontsize=10, rotation=25)
plt.yticks(fontsize=10)
plt.show()
```

> Dapat ditarik dari visualisasi total pengeluaran konsumen per bulan pada tahun `2018`. Pada grafik tersebut, terlihat bahwa puncak pengeluaran terjadi pada bulan `November`, sedangkan pengeluaran paling rendah terjadi pada bulan `September`.
